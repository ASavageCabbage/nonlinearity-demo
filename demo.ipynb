```python
import numpy as np
import matplotlib.pyplot as plt

DATAPOINTS = 1000
timestamps = np.arange(0, DATAPOINTS)

STEPS = 5
JITTER = 0.1
SLOPE = 2
POWER = 1.7
x_data = np.concatenate([np.random.normal(loc=(i+1)/JITTER, size=DATAPOINTS//STEPS) for i in range(STEPS)])
y_data = SLOPE * x_data ** POWER + np.random.normal(size=x_data.shape)
timestamps = timestamps[:len(x_data)]

fig, axs = plt.subplots(2, 1, figsize=(6,4), sharex=True)
x_ax, y_ax = axs
x_ax.plot(timestamps, x_data, "r.")
x_ax.set_ylabel("foo")
y_ax.plot(timestamps, y_data, "b.")
y_ax.set_ylabel("bar")
y_ax.set_xlabel("time")
for ax in axs:
    ax.set_ylim(0, None)
fig.tight_layout()

fig, ax = plt.subplots(1, 1, figsize=(4,4))
ax.plot(x_data, y_data, "k.")
ax.set_xlabel("foo")
ax.set_xlim(0, None)
ax.set_ylabel("bar")
ax.set_ylim(0, None)
fig.tight_layout()
```

```python
DATAPOINTS = 1000
timestamps = np.arange(0, DATAPOINTS)

AMPLITUDE = 10
PERIODS = 2
OFFSET = 2
JITTER = 0.1
SLOPE = 2
POWER = 1.7
x_data = AMPLITUDE * (
    np.cos(np.linspace(0, 2*PERIODS*np.pi, DATAPOINTS))
    + np.random.normal(scale=JITTER, size=timestamps.shape)
    + OFFSET
)
y_data = SLOPE * x_data ** POWER + np.random.normal(scale=SLOPE*JITTER, size=x_data.shape)

fig, axs = plt.subplots(2, 1, figsize=(6,4), sharex=True)
x_ax, y_ax = axs
x_ax.plot(timestamps, x_data, "r.")
x_ax.set_ylabel("foo")
y_ax.plot(timestamps, y_data, "b.")
y_ax.set_ylabel("bar")
y_ax.set_xlabel("time")
for ax in axs:
    ax.set_ylim(0, None)
fig.tight_layout()

fig, ax = plt.subplots(1, 1, figsize=(4,4))
ax.plot(x_data, y_data, "k.")
ax.set_xlabel("foo")
ax.set_xlim(0, None)
ax.set_ylabel("bar")
ax.set_ylim(0, None)
fig.tight_layout()
```

```python
DATAPOINTS = 25
START = 10
JITTER = 0.1
SLOPE = 2
POWER = 1

x_data = np.arange(START, START + DATAPOINTS) + np.random.normal(scale=JITTER, size=DATAPOINTS)
y_data = SLOPE * x_data ** POWER + np.random.normal(scale=SLOPE*JITTER, size=DATAPOINTS)

SAMPLES = 3
lines = []
for i in range(DATAPOINTS % SAMPLES, DATAPOINTS, DATAPOINTS // SAMPLES):
    fig, ax = plt.subplots(1, 1, figsize=(4,4))
    ax.plot(x_data, y_data, "k.")
    lines.append(([0, x_data[i]], [0, y_data[i]]))
    for x, y in lines:
        ax.plot(x, y, "r.-")
```

```python
slope_per_point = y_data / x_data
fig, ax = plt.subplots(1, 1, figsize=(6,4))
ax.plot(slope_per_point, "k.")
ax.set_xlabel("Data index")
ax.set_ylabel("Slope to origin ($y/x$)")
```
